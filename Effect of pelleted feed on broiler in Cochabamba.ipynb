## 3.3 Statistical model.


<p style="text-align: justify;">The statistical models used to identify treatments with the best response were as follows:</p>


Yij = μ + Φi + Τj + εij

i= 1, 2, 3, 4, r blocks.

j= 1, 2, 3, 4, t pelleted feed.

Y= Response variable observed in a pen of the i-th block where the j-th pelleted feed was applied.

μ = Overall mean.

Φi = Random effect of the i-th block.

Τj = Fixed effect of the j-th pelleted feed.

εij= Random effect of residuals.

Significance level α=0.05

<p style="text-align: justify;">The equation to be presented is based on the general equation of Yij = μ + Φi + Τj + εij, where i represents the blocks, with a total of r blocks, and j represents the different types of feed, with a total of t types of pelleted feed. The response variable Y is the observed variable in a pen of the i-th block where the j-th pelleted feed was applied. The term μ represents the overall mean of all blocks and types of pelleted feed. The term Φi represents the random effect of the i-th block and the term Τj represents the fixed effect of the j-th pelleted feed. The term εij represents the random error or residual associated with the i-th pen and j-th pelleted feed type. The significance level α used in the study was 0.05. This report aims to describe the results obtained from the analysis of the collected data and provide a detailed understanding of the relationship between the variables involved in the study.</p>

<p style="text-align: justify;">This code snippet uses the shapiro function from the Scipy library to perform a normality test on the response variable weight gain. The test results are printed, and a histogram of the data distribution is generated.</p>

### 4.2.1.2 Analysis of variance.

<p style="text-align: justify;">The presented code is an implementation of one-way Analysis of Variance (ANOVA) using the SciPy library in Python. The purpose of ANOVA is to analyze the variation of a variable of interest in relation to a categorical variable, in this case, the variables "Block" and "treatment" in relation to the "Weight Gain" variable.</p>

```python
import scipy.stats as stats

# select relevant columns from the dataframe
df_selected = df[["Block", "treatment", "WG Day 49"]]

# perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(*[group["WG Day 49"] for name, group in df_selected.groupby(["Block", "treatment"])])

# create results table
anova_table1 = pd.DataFrame({'FV': ['Treatment', 'Block', 'Error', 'Total'],
                            'GL': [3, 1, 12, 15],
                            'CM': ['-','-','-','-']},
                           columns=['FV', 'GL', 'CM'])

anova_table1.loc[0,'CM'] = round(sum(df_selected.groupby('treatment').sum()['WG Day 49']**2)/4 - (df_selected['WG Day 49'].sum()**2)/16, 2)
anova_table1.loc[1,'CM'] = round(sum(df_selected.groupby('Block').sum()['WG Day 49']**2)/2 - (df_selected['WG Day 49'].sum()**2)/16, 2)
anova_table1.loc[2,'CM'] = round(anova_table1.loc[0,'CM'] + anova_table1.loc[1,'CM'], 2)
anova_table1.loc[3,'CM'] = round(sum(df_selected['WG Day 49']**2) - (df_selected['WG Day 49'].sum()**2)/16, 2)

anova_table1['F cal'] = ['-', round(anova_table1.loc[1,'CM']/anova_table1.loc[2,'CM'], 2), round(f_statistic, 2), '-']
anova_table1['Pr>F'] = ['-', round(stats.f.sf(anova_table1.loc[1,'F cal'], 1, 12), 4), round(p_value, 4), '-']

# calculate coefficient of variation and add a new column to the results table
cv = round(df_selected["WG Day 49"].std() / df_selected["WG Day 49"].mean() * 100, 2)
anova_table1['CV %'] = ['-', '-', cv, '-']

anova_table1

```

<p style="text-align: justify;">The results of the analysis of variance indicate that there is a statistically significant difference between treatments, as the p-value (Pr>F) for the "Treatment" factor is less than 0.05, meaning that there is enough evidence to reject the null hypothesis of no significant difference between treatments. However, the "Block" factor does not show a significant difference, as its p-value is greater than 0.05. Additionally, the coefficient of variation (CV%) calculated for the error is 5.15%, suggesting moderate variability in the data that is not explained by the factors included in the model.</p>

<p style="text-align: justify;">In summary, the results indicate that treatment has a significant effect on the response variable, but block does not have a significant effect. Additionally, there is moderate variability in the data that is not explained by the factors included in the model.</p>

### 4.2.1.3 Hypothesis testing and comparison of means.

<p style="text-align: justify;">The following is an ANOVA analysis to determine if there are significant differences between the means of the four treatment groups. The f_oneway function calculates the F-value and p-value.</p>

<p style="text-align: justify;">Finally, a Tukey test is performed to compare all possible pairs of groups and determine if there are significant differences between the means of these pairs. The results of the Tukey test are printed on the screen.</p>

<p style="text-align: justify;">The Python code used was useful for performing a statistical analysis of the data in a dataframe, which allows determining if there are significant differences between the means of the treatment groups and comparing these means among themselves.</p>

```python
import statsmodels.stats.multicomp as mc
from scipy.stats import f_oneway
import seaborn as sns

# Data selection
data = df[['treatment', 'WG Day 49']]

# Separation of data by treatment
treatment1 = data[data['treatment'] == 'T1']['WG Day 49']
treatment2 = data[data['treatment'] == 'T2']['WG Day 49']
treatment3 = data[data['treatment'] == 'T3']['WG Day 49']
treatment4 = data[data['treatment'] == 'T4']['WG Day 49']

# Analysis of variance
fvalue, pvalue = f_oneway(treatment1, treatment2, treatment3, treatment4)

print("F Value: ", fvalue)
print("p Value: ", pvalue)

# Perform Tukey's test
comp = mc.MultiComparison(data['WG Day 49'], data['treatment'])
tukey_results = comp.tukeyhsd(alpha=0.05)

print(tukey_results)


# Generate bar graph
fig, ax = plt.subplots(figsize=(8,6))
sns.barplot(data=data, x='treatment', y='WG Day 49', errorbar="sd", capsize=.2)

ax.set_xlabel("Treatment")
ax.set_ylabel("Weight Gain (g)")
ax.set_ylim([2400, 2900])

plt.show()
```

<p style="text-align: justify;">The analysis of variance (ANOVA) found a statistically significant difference among the groups (F-value = 12.22, p = 0.0006). To determine which groups are significantly different, a multiple comparison analysis of means (Tukey HSD) was performed.</p>

<p style="text-align: justify;">The results indicate that there are significant differences in the means between treatments T1, T3, and T4. However, there are no significant differences between treatments T1 and T2, nor between treatments T2 and T4. Treatments T3 and T4 also do not differ significantly from each other.</p>

<p style="text-align: justify;">In summary, the results suggest that treatments T1, T3, and T4 differ significantly from each other in terms of weight gain, while treatment T2 does not differ significantly from the other treatments.</p>

### 4.2.2.3 Hypothesis testing and comparison of means.

<p style="text-align: justify;">In the same way, with the python code, a Tukey test is performed to compare all possible pairs of groups and determine if there are significant differences in mortality between these pairs.</p>

```python
# Data selection
data = df[['treatment', 'M Day 49']]

# Separation of data by treatment
treatment1 = data[data['treatment'] == 'T1']['M Day 49']
treatment2 = data[data['treatment'] == 'T2']['M Day 49']
treatment3 = data[data['treatment'] == 'T3']['M Day 49']
treatment4 = data[data['treatment'] == 'T4']['M Day 49']

# Analysis of variance
fvalue, pvalue = f_oneway(treatment1, treatment2, treatment3, treatment4)

print("F Value: ", fvalue)
print("p Value: ", pvalue)

# Perform Tukey's test
comp = mc.MultiComparison(data['M Day 49'], data['treatment'])
tukey_results = comp.tukeyhsd(alpha=0.05)

print(tukey_results)


# Generate bar graph
fig, ax = plt.subplots(figsize=(8,6))
sns.barplot(data=data, x='treatment', y='M Day 49', errorbar="sd", capsize=.2)

ax.set_xlabel("Treatment")
ax.set_ylabel("Mortality %")
ax.set_ylim([0, 15])

plt.show()
```

<p style="text-align: justify;">The "meandiff" column shows the difference between the means of each pair of groups, and the "p-adj" column shows the adjusted p-value for each pairwise comparison. The "lower" and "upper" columns represent the lower and upper limits of the confidence interval, respectively. If "reject" is "True", it means that the null hypothesis of equality of means for that pair of treatments is rejected.</p>

<p style="text-align: justify;">In this case, none of the pairwise comparisons show a significant difference between the means, since all adjusted p values are greater than 0.05. Therefore, we cannot conclude that either treatment has a significantly different effect on mortality.</p>

The results are then compiled into a results table, which includes the sources of variation, degrees of freedom, root mean squares, F-values, and p-values. Also, the coefficient of variation is calculated and added to the table to give an idea of the variability of the data.</p>

```python
# select relevant columns from the dataframe
df_selected3 = df[["Block", "treatment", "FCR Day 49"]]

# perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(*[group["FCR Day 49"] for name, group in df_selected3.groupby(["Block", "treatment"])])

# create results table
anova_table3 = pd.DataFrame({'FV': ['Treatment', 'Block', 'Error', 'Total'],
                            'GL': [3, 1, 12, 15],
                            'CM': ['-','-','-','-']},
                           columns=['FV', 'GL', 'CM'])

anova_table3.loc[0,'CM'] = round(sum(df_selected3.groupby('treatment').sum()['FCR Day 49']**2)/4 - (df_selected3['FCR Day 49'].sum()**2)/16, 2)
anova_table3.loc[1,'CM'] = round(sum(df_selected3.groupby('Block').sum()['FCR Day 49']**2)/2 - (df_selected3['FCR Day 49'].sum()**2)/16, 2)
anova_table3.loc[2,'CM'] = round(anova_table3.loc[0,'CM'] + anova_table3.loc[1,'CM'], 2)
anova_table3.loc[3,'CM'] = round(sum(df_selected3['FCR Day 49']**2) - (df_selected3['FCR Day 49'].sum()**2)/16, 2)

anova_table3['F cal'] = ['-', round(anova_table3.loc[1,'CM']/anova_table3.loc[2,'CM'], 2), round(f_statistic, 2), '-']
anova_table3['Pr>F'] = ['-', round(stats.f.sf(anova_table3.loc[1,'F cal'], 1, 12), 4), round(p_value, 4), '-']

# calculate coefficient of variation and add a new column to the results table
cv = round(df_selected3["FCR Day 49"].std() / df_selected3["FCR Day 49"].mean() * 100, 2)
anova_table3['CV %'] = ['-', '-', cv, '-']

anova_table3
```

<p style="text-align: justify;">The calculated F value for the block is 1.0, indicating that the variation between blocks is similar to the variation within blocks. Furthermore, the p-value of 0.337 indicates that there is not enough evidence to reject the null hypothesis that there are no significant differences between the blocks. On the other hand, the MC value for the error is greater than the MC for the treatment and the block, which suggests that the greatest source of variation is found within the treatments. The CV of 5.98% indicates that there is moderate variability in the data.</p>

### 4.2.3.3 Hypothesis testing and comparison of means.

<p style="text-align: justify;">In this code, an analysis is performed on the Feed Conversion Rate (FCR) variable by separating the data by treatment and applying an analysis of variance (ANOVA) to determine if there are significant differences between treatments. Then, a Tukey test is performed to identify which treatments show significant differences from each other. In addition, a bar graph is generated to visualize the differences in FCR between treatments.</p>

```python
# Data selection
data = df[['treatment', 'FCR Day 49']]

# Separation of data by treatment
treatment1 = data[data['treatment'] == 'T1']['FCR Day 49']
treatment2 = data[data['treatment'] == 'T2']['FCR Day 49']
treatment3 = data[data['treatment'] == 'T3']['FCR Day 49']
treatment4 = data[data['treatment'] == 'T4']['FCR Day 49']

# Analysis of variance
fvalue, pvalue = f_oneway(treatment1, treatment2, treatment3, treatment4)

print("F Value: ", fvalue)
print("p Value: ", pvalue)

# Perform Tukey's test
comp = mc.MultiComparison(data['FCR Day 49'], data['treatment'])
tukey_results = comp.tukeyhsd(alpha=0.05)

print(tukey_results)


# Generate bar graph
fig, ax = plt.subplots(figsize=(8,6))
sns.barplot(data=data, x='treatment', y='FCR Day 49', errorbar="sd", capsize=.2)

ax.set_xlabel("Treatment")
ax.set_ylabel("Feed Conversion Rate")
ax.set_ylim([1.8, 2.3])

plt.show()
```

<p style="text-align: justify;">Based on the results obtained for the FCR response variable, an analysis of variance (ANOVA) and a Tukey test were performed to compare the means of the different treatments. The F value obtained was 1.397, with a p value of 0.2914, which indicates that there are no significant differences between the means of the treatments.</p>

<p style="text-align: justify;">Tukey's test also shows that there are no significant differences between the treatments, since no pair of treatments showed a significant difference in their means.</p>

<p style="text-align: justify;">Therefore, we can conclude that there are no significant differences in the Feed Conversion Rate (FCR) between the different treatments.</p>
